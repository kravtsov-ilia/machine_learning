## Часть №2. Моделирование при помощи kNN.


<b>Разобёем выборку на признаки и целевую переменную</b>

```python
X = df_cut.drop(columns='diagnosis')
y = df_cut['diagnosis'].map({'M': 1, 'B': 0})
```

<b>Разобьем выборку на train-test</b>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<b>Как мы можем заметить из части №1, все признаки имеют вещественный тип и лежат в разных диапазонах,
поэтому нормализуем их, чтобы все они давали равнозначный результат при поиске расстояний между точками</b>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

<b>Обучим нашу модель классификатором "из коробки" на масштабированных данных</b>

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
```

<b>Оценим точность получившейся модели</b>

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
accuracy_train = accuracy_score(y_pred=knn.predict(X_train), y_true=y_train)
accuracy_test = accuracy_score(y_pred=knn.predict(X_test), y_true=y_test)
```

```python
precision_train = precision_score(y_pred=knn.predict(X_train), y_true=y_train)
```

```python
precision_test = precision_score(y_pred=knn.predict(X_test), y_true=y_test)
```

```python
recall_train = recall_score(y_pred=knn.predict(X_train), y_true=y_train)
```

```python
recall_test = recall_score(y_pred=knn.predict(X_test), y_true=y_test)
```

```python
f1_train = f1_score(y_pred=knn.predict(X_train), y_true=y_train)
```

```python
f1_test = f1_score(y_pred=knn.predict(X_test), y_true=y_test)
```

```python
accuracy_train, precision_train, recall_train, f1_train
```

```python
accuracy_test, precision_test, recall_test, f1_test
```

<b>Как можно заметить, получивщаяся модель имеет достаточно низкую точность предсказания и не годиться для реального применения</b>
<b>Также из оценки параметров точности предсказания видно, что наша модель не предсказывает ложно отрицательные результаты</b> <i>(что в принципе 
для предсказания злокачественной опухоли, наверное, не так уж и плохо ? интересно как так получилось?)</i>

</br>

<b>Построим ROC-кривую и посчитаем площадь под ней</b>

```python
from sklearn.metrics import roc_curve, auc
```

```python
fpr, tpr, threshold = roc_curve(y_test, knn.predict(X_test))
roc_auc = auc(fpr, tpr)
```

```python
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC Curve of kNN')
plt.show()
```

<b>AUC - равен 0,5 что ещё раз говорит нам о непригодности данной модели для предсказания</b>

<b>Попробуем найти оптимальное число соседей используя кросс-валидацию, 
    наиболее качественной будем считать модель с наибольшей точностью (precision) 
    так как для постановки диагноза важна доля ложно-отрицательных результатов</b>

```python
from sklearn.model_selection import cross_val_score
neighbors = range(1, 50)
prec_list = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=5, scoring='precision')
    prec_list.append((k, scores.mean()))
    
```

```python
prec_list.sort(key=lambda x: x[1], reverse=True)
prec_list
```

```python
best_neighbors_cnt = prec_list[0][0]
best_neighbors_cnt
```

```python
knn = KNeighborsClassifier(n_neighbors=best_neighbors_cnt)
knn.fit(X_train, y_train)
fpr, tpr, _ = roc_curve(y_test, knn.predict(X_test))
roc_auc = auc(fpr, tpr)
```

```python
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC Curve of kNN')
plt.show()
```

<b>Модель стала лучше</b>  
<b>Попробуем запустить все то же самое, только на масштабированных данных</b>

```python
neighbors = range(1, 50)
prec_list = []
X_all_scaled = np.concatenate((X_train_scaled, X_test_scaled))
y_all_scaled = np.concatenate((y_train, y_test))
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_all_scaled, y_all_scaled, cv=5, scoring='precision')
    prec_list.append((k, scores.mean()))

prec_list.sort(key=lambda x: x[1], reverse=True)

best_neighbors_cnt = prec_list[0][0]
print(f'best_neighbors_cnt: {best_neighbors_cnt}')

knn = KNeighborsClassifier(n_neighbors=best_neighbors_cnt)
knn.fit(X_train, y_train)
fpr, tpr, _ = roc_curve(y_test, knn.predict(X_test))
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC Curve of kNN')
plt.show()
```

<b>Результат не сильно изменился, даже немного стал хуже</b>

<b>В целом можно сказать, что алгоритм "из коробки" лучше не использовать, лучше всегда подбирать число соседей</b>

# Попробуем посмотреть каких результатов можно добиться используя логистическую регрессию

#### Коррелирующие признаки мы уже убрали из модели

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
def quality_report(prediction, actual):
    print("Accuracy: {:.3f}\nPrecision: {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}".format(
        accuracy_score(prediction, actual),
        precision_score(prediction, actual),
        recall_score(prediction, actual),
        f1_score(prediction, actual)
    ))
```

```python
quality_report(log_reg.predict(X_train_scaled), y_train)
```

```python
quality_report(log_reg.predict(X_test_scaled), y_test)
```

### Построим ROC-кривую

```python
fpr, tpr, _ = roc_curve(y_test, log_reg.predict(X_test_scaled))
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC Curve of Logistic Regression')
plt.show()
```

### Посмотрим на влияние признаков

```python
featureImportance = pd.DataFrame({"feature": df_cut.drop('diagnosis', axis=1).columns, 
                                  "importance": log_reg.coef_[0]})
featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot.bar();
```

### Попробуем найти наилучший коэффициент регурялизации С

```python
from sklearn.linear_model import LogisticRegressionCV
best_C = None
best_fit = None
best_roc_auc = 0
for i in range(14):
    C = i + 1
    log_reg = LogisticRegressionCV(Cs=C,cv=5, random_state=1)
    log_reg.fit(X_train_scaled, y_train)
    fpr, tpr, _ = roc_curve(y_test, log_reg.predict(X_test_scaled))
    roc_auc = auc(fpr, tpr)
    if roc_auc > best_roc_auc:
        best_C = C
        best_roc_auc = roc_auc
        best_fit = log_reg
```

```python
print(f'best_C: {best_C}')
print(f'best_roc_auc: {best_roc_auc}')
quality_report(best_fit.predict(X_test_scaled), y_test)
```
