```python
sns.heatmap(corr, annot=True, fmt='.2f');
```

```python
# Найдмем пары признаков с корреляцией больше 0.6
```

```python
CorField = []
MainField = []
for i in corr:
    for j in corr.index[corr[i] > 0.7]:
        if i != j and j not in CorField and i not in CorField:
            CorField.append(j)
            MainField.append(i)
            print (i, j, corr[i][corr.index == j].values[0])
```
